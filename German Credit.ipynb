## Project Title:
#### Credit Risk Analysis: Identifying High-Risk Borrowers and Reducing Default Risk

## Project Overview
In consumer lending, credit risk assessment plays a critical role in minimizing financial losses and ensuring sustainable portfolio growth. Banks must evaluate whether a loan applicant is likely to repay their debt based on personal, financial, and credit-related characteristics.
This project uses the German Credit dataset to analyze borrower behavior and identify factors associated with high-risk (default-prone) customers.
### Project Objective
The primary objective of this project is to:
- Identify borrower profiles that are most likely to be classified as high credit risk
- Understand the key drivers of credit risk using interpretable models
- Provide actionable recommendations for banks to reduce potential bad debt

## Business Questions
1. Which customers are most likely to be high-risk borrowers?
2. How can banks proactively reduce bad debt risk?
3. What are the key drivers of credit risk using interpretable models?
