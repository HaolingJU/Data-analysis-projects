## Project Overview
This project focuses on exploratory and business-oriented analysis to identify key
risk drivers and actionable insights. While predictive modeling can further support
decision-making, the emphasis here is on interpretability and understanding risk patterns
that are directly applicable to credit policy design.

## Business Questions

1. Which customers are most likely to be high-risk borrowers?
2. How can banks proactively reduce bad debt risk?
3. What are the key drivers of credit risk using interpretable models?

## 1. Data Understanding

### Summary

- The dataset contains **1,000 loan applicants** with **10 features**.
- Feature composition:
  - **6 categorical variables**
  - **4 numerical variables**
- Average borrower age: **~35.5 years**

**Credit Amount**
- Median ≈ 2,319
- Maximum ≈ 18,424 (indicating a strong right-skewed distribution)

**Loan Duration**
- Minimum: 4 months
- Maximum: 72 months

These characteristics suggest substantial heterogeneity in both borrower profiles and loan
structures.

---

## 2. Data Cleaning and Preparation

- The target variable `Risk` is converted into a binary label:
  - `0` = Good credit
  - `1` = Bad credit
- Missing values are primarily observed in:
  - `Saving accounts`
  - `Checking account`
- Instead of imputing with the most frequent category, missing values are labeled as
  **`unknown`** to preserve information and avoid introducing bias.

### 3.2 Borrower Profile Analysis

#### Bad Credit Rate by Age Group

Younger borrowers exhibit higher default rates, particularly those aged **18–25**.
Default risk decreases steadily with age, suggesting that life-stage stability plays an
important role in repayment behavior.

![Bad Credit Rate by Age Group](figures/eda/bad_rate_by_age_group.png)

---

#### Bad Credit Rate by Housing Status

Borrowers without stable housing ownership (renting or free housing) show higher bad credit
rates compared to homeowners. Housing ownership appears to be a strong proxy for financial
stability.

![Bad Credit Rate by Housing Status](figures/eda/bad_credit_rate_by_housing.png)

---

#### Bad Credit Rate by Employment Category

Employment stability also differentiates credit risk. Unemployed borrowers or those with
less stable job categories exhibit higher default rates.

![Bad Credit Rate by Job Category](figures/eda/bad_credit_rate_by_job.png)

### 3.3 Loan Characteristics and Credit Risk

#### Loan Duration Distribution by Credit Risk

Loan duration shows a strong association with default risk. Bad credit borrowers tend to
have longer loan terms, with significantly higher maximum durations.

![Loan Duration Distribution by Credit Risk](figures/eda/duration_distribution_by_risk.png)

---

#### Bad Credit Rate by Loan Duration Group

Default rates increase monotonically as loan duration becomes longer. Long-term loans
(37–72 months) represent a particularly high-risk segment.

![Bad Credit Rate by Loan Duration Group](figures/eda/bad_rate_by_duration_group.png)

---

### 3.4 Loan Purpose and Interaction Effects

#### Bad Credit Rate by Loan Purpose

Credit risk varies meaningfully across loan purposes. Business- and education-related loans
exhibit higher default rates compared to consumer-oriented purposes.

![Bad Credit Rate by Loan Purpose](figures/eda/bad_rate_by_purpose.png)

---

#### Interaction Analysis: Loan Duration × Loan Purpose

Risk escalates sharply when **long loan durations** are combined with **high-uncertainty loan
purposes**. Long-term business and education loans emerge as the highest-risk segments.

![Interaction: Loan Duration × Loan Purpose](figures/eda/interaction_duration_purpose.png)

---

### EDA Summary

Overall, credit risk is driven by a combination of borrower stability and loan structure.
Among all factors, **loan duration** stands out as the most influential and controllable
risk driver, especially when interacting with loan purpose.


## 4. Business Question Analysis

### 4.1 Which customers are most likely to be high-risk borrowers?

High-risk borrowers are more likely to combine multiple instability factors, including:

- Younger age
- Unstable employment status
- Non-homeownership

These combined factors provide stronger risk signals than any single demographic attribute
alone.

---

### 4.2 How can banks proactively reduce bad debt risk?

Key risk drivers are not purely borrower-related but are also embedded in **loan design
choices**, many of which are controllable by banks.

Recommended risk mitigation strategies include:

- **Actively managing loan duration**
  Longer loan terms are consistently associated with higher default risk. Stricter approval
  criteria or shorter maximum terms can reduce exposure.

- **Setting risk-sensitive credit amount thresholds**
  Very large loans exhibit elevated default risk and greater volatility. Tiered limits and
  enhanced reviews can help mitigate this risk.

- **Differentiating policies by loan purpose**
  Business and education loans require stricter underwriting and monitoring compared to
  lower-risk consumer loans.

- **Incorporating interaction effects into decision-making**
  Long-duration loans for high-uncertainty purposes should be treated as priority risk
  segments.
